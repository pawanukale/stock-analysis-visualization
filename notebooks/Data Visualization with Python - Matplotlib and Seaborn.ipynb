```python
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
cancer
```

```python
df_cancer= pd.DataFrame(np.c_[cancer['data'], cancer['target']], columns = np.append(cancer['feature_names'],['target']))
```

```python
df_cancer
```

```python
df_cancer.head()
```

```python
df_cancer.tail()
```

```python
sns.scatterplot(x='mean area',y='mean smoothness',hue ='target',data=df_cancer)
```

```python
plt.figure(figsize=(10,10))
sns.countplot(df_cancer['target'],label='Count')
```

```python
sns.scatterplot(x='mean radius',y='mean area', hue='target',data=df_cancer)
```

```python
sns.pairplot(df_cancer,hue='target',vars =['mean radius','mean texture','mean perimeter','mean smoothness'])
```

```python
plt.figure(figsize=(30,20))
sns.heatmap(df_cancer.corr(),annot=True)
```

```python
sns.distplot(df_cancer['mean radius'], bins=25, color='b')
```
